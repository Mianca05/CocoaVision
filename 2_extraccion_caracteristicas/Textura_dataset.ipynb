```python
"""Librerias"""
"""tratar y extraer caracteristicas de imagenes"""
import cv2
"""manipular rutas del host"""
import os
"""operaciones con arrays"""
import numpy as np
"""tratar valores atipicos"""
from sklearn.preprocessing import MinMaxScaler
"""almacenar grandes volumenes de datos"""
import h5py
"""encontrar directorios"""
import pathlib
"""procesar y manipular imagenes"""
import mahotas as mt
```

# Extraer caracteristicas de textura de imagen

```python
"""obtener directorio de imagenes"""
dir_imgda="../EnfermedadesCacao_DA/"
"""comprobar directorio principal de imagenes"""
img_dir_da = pathlib.Path(dir_imgda)
"""obtener clases de las imagenes"""
clases_img2 = os.listdir(img_dir_da)

"""crear diccionario para el acceso de imagenes"""
cacao_images_dict = {
    'monilia': list(img_dir_da.glob('Monilia/*')),
    'sana': list(img_dir_da.glob('Sana/*'))
}

"""crear diccionario de etiquetas"""
cacao_etiquetas = {
    'monilia': 0,
    'sana':1
}
"""mostrar clases y directorio local"""
print(clases_img2)
img_dir_da
```

```python
"""crear listas para almacenar caracteristicas de textura y su etiqueta"""
caracteristicas_textura = []
etiquetas_caract_textura   = []


"""funcion para extraer la textura de una imagen"""
def extraer_caracteristicas_textura(image):
    """obtener texturas de la imagen"""
    textures = mt.features.haralick(image)
    """calcular el promedio de la textura de la imagen por columna"""
    htextura_mean  = textures.mean(axis=0)
    return htextura_mean
```

```python
"""seleccionar clase e imagenes a partir del diccionario de imgs""" 
for clase, imgs_cacao in cacao_images_dict.items():
    """mostrar clase cacao"""
    print('Clase:',clase)
    i = 1
    """recorrer el conjunto de imagenes"""
    for imagen in range(len(imgs_cacao)):
        """obtener img"""
        d_img = cv2.imread(str(cacao_images_dict[clase][imagen]))
        """convertir imagen en escala de grises"""
        grises = cv2.cvtColor(d_img, cv2.COLOR_BGR2GRAY)
        """extraer textura"""
        caracteristica_textura = extraer_caracteristicas_textura(grises)
        """agregar datos de las caracteristicas de grises a las listas"""
        caracteristicas_textura.append(caracteristica_textura)
        etiquetas_caract_textura.append(clase)
        print(f"Imagen: {i}")
        i+=1
```

# Crear dataset para textura

```python
"""mostrar tamanio final del vector de caracteristicas de textura"""
print(f"Caracteristicas de colores: {np.array(caracteristicas_textura).shape}")
"""mostrar tamanio final del vector de etiquetas de caracteristicas de textura"""
print(f"Etiqueta de caracteristicas de colores: {np.array(etiquetas_caract_textura).shape}")
```

```python
"""codificar etiquetas"""
def replace(lista, dicc):
    """retornar lista con etiquetas"""
    return [x if x not in dicc else dicc[x] for x in lista]
target = replace(etiquetas_caract_textura, cacao_etiquetas)

"""instanciar clase de escalamiento por minimos y maximos"""
scaler = MinMaxScaler(feature_range=(0, 1))
"""normalizar vector"""
rescalar_caracteristicas_textura = scaler.fit_transform(caracteristicas_textura)
```

```python
"guardar vector de caracteristicas de textura con HDF5"
h5f_data = h5py.File('/home/miguel/Documents/ProyectoMoniliaCacao/2BDD_Caracteristicas/textura_caracteristicas.h5', 'w')
h5f_data.create_dataset('Cacao_Textura_Caracteristicas', data=np.array(rescalar_caracteristicas_textura))

"guardar vector de caracteristicas de textura con HDF5"
h5f_label = h5py.File('/home/miguel/Documents/ProyectoMoniliaCacao/2BDD_Caracteristicas/textural_etiquetas.h5', 'w')
h5f_label.create_dataset('Cacao_Textura_Etiquetas', data=np.array(target))

h5f_data.close()
h5f_label.close()
```
