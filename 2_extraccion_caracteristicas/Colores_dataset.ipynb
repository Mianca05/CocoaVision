```python
"""Librerias"""
"""tratar y extraer caracteristicas de imagenes"""
import cv2
"""manipular rutas del host"""
import os
"""operaciones con arrays"""
import numpy as np
"""tratar valores atipicos"""
from sklearn.preprocessing import MinMaxScaler
"""almacenar grandes volumenes de datos"""
import h5py
"""encontrar directorios"""
import pathlib
```

# Extraer caracteristicas de color 

```python
"""obtener directorio de imagenes"""
dir_imgda="../EnfermedadesCacao_DA/"
"""comprobar directorio principal de imagenes"""
img_dir_da = pathlib.Path(dir_imgda)
"""obtener clases de las imagenes"""
clases_img2 = os.listdir(img_dir_da)

"""crear diccionario para el acceso de imagenes"""
cacao_images_dict = {
    'monilia': list(img_dir_da.glob('Monilia/*')),
    'sana': list(img_dir_da.glob('Sana/*'))
}

"""crear diccionario de etiquetas"""
cacao_etiquetas = {
    'monilia': 0,
    'sana':1
}
"""mostrar clases y directorio local"""
print(clases_img2)
img_dir_da
```

```python
"""crear histograma de colores para describir caracteristicas"""
def color_histogram(image, mask=None):
    """cambiar la imagen en modelo de color HSV"""
    image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
    """analizar el histograma""" 
    histograma  = cv2.calcHist([image], [0, 1, 2], None, [bins, bins, bins], [0, 256, 0, 256, 0, 256])
    """normalizacion del histograma"""
    cv2.normalize(histograma, histograma)
    return histograma.flatten()
```

```python
"""crear listas para almacenar caracteristicas de colores"""
caracteristicas_colores = []
etiquetas_caract_colores   = []
```

```python
"""seleccionar clase e imagenes a partir del diccionario de imgs""" 
for clase, imgs_cacao in cacao_images_dict.items():
    """mostrar clase cacao"""
    print('Clase:',clase)
    i = 1
    """recorrer el conjunto de imagenes"""
    for imagen in range(len(imgs_cacao)):
        """obtener img"""
        d_img = cv2.imread(str(cacao_images_dict[clase][imagen]))
        "extrar color de la imagen leida en el histograma de colores"
        bins = 8
        histograma = color_histogram(d_img)        
        """agregar datos de las caraccteristicas a las listas"""
        caracteristicas_colores.append(histograma)
        etiquetas_caract_colores.append(clase)
        print(f"Imagen: {i}")
        i+=1
```

# Crear dataset

```python
"""mostrar tamanio final del vector de caracteristicas de color"""
print(f"Caracteristicas de colores: {np.array(caracteristicas_colores).shape}")
"""mostrar tamanio final del vector de color de etiquetas de caracteristicas de color"""
print(f"Etiqueta de caracteristicas de colores: {np.array(etiquetas_caract_colores).shape}")
```

```python
"""codificar etiquetas"""
def replace(lista, dicc):
    return [x if x not in dicc else dicc[x] for x in lista]
target = replace(etiquetas_caract_colores, cacao_etiquetas)
```

```python
"""instanciar escalamiento de minimos y maximos"""
escalar    = MinMaxScaler(feature_range=(0, 1))
"""normalizar vector"""
rescalar_caracteristicas_color = escalar.fit_transform(caracteristicas_colores)
```

```python
"guardar vector de caracteristicas de colores con HDF5"
h5f_data = h5py.File('/home/miguel/Documents/ProyectoMoniliaCacao/2BDD_Caracteristicas/color_caracteristicas.h5', 'w')
h5f_data.create_dataset('Cacao_Color_Caracteristicas', data=np.array(rescalar_caracteristicas_color))
"""guardar vector de colores"""
h5f_label = h5py.File('/home/miguel/Documents/ProyectoMoniliaCacao/2BDD_Caracteristicas/color_etiquetas.h5', 'w')
h5f_label.create_dataset('Cacao_Color_Etiquetas', data=np.array(target))
"""cerrar archivo"""
h5f_data.close()
h5f_label.close()
```
